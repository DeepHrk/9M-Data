# Data Preprocessing

```python
# Handle Null Values
data.fillna("Unknown", inplace=True)
```

```python
#The column name have spaces we can fill the space with '_'
data = data.rename(columns={'Unnamed: 0':'Unnamed0', 'year founded':'year_founded',
       'size range':'size_range', 'linkedin url':'linkedin_url',
       'current employee estimate':'current_employee_estimate', 'total employee estimate':'total_employee_estimate'})
```

```python
# Drop Unnamed0 column
data.drop('Unnamed0', axis=1, inplace=True)
```

```python
def find_companies_by_industry_and_location_and_size_range(df, industry_req, location_req, size_range_req):
    # Clean up industry and location and size_range input to remove extra spaces
    industry_req = industry_req.strip().lower()
    location_req = location_req.strip().lower()
    size_range_req = size_range_req.strip().lower()

    # List to store all matching companies
    matching_companies = []

    # Loop through each row in the DataFrame
    for idx, row in df.iterrows():
        # Check if the industry and location and size_range  match the user requirements
        industry = str(row['industry']).strip()
        location = str(row['locality']).strip()
        size_range = str(row['size_range']).strip()

        # Ensure both fields are not NaN and compare
        if pd.notna(industry) and pd.notna(location) and pd.notna(size_range):  # Avoid NaN values
            if industry_req in industry.lower() and location_req in location.lower() and size_range_req in size_range.lower():
                matching_companies.append(row)
    
    # Convert the list of matching companies to a DataFrame and return it
    return pd.DataFrame(matching_companies)

# Get user input
industry_req = input("Enter the industry you're searching for: ")  
location_req = input("Enter the location you're searching for: ") 
size_range_req = input("Enter The size range you're searching for e.g('10001+', '5001 - 10000', '1001 - 5000', '501 - 1000', '201 - 500','51 - 200', '11 - 50', '1 - 10')")

# Assuming 'data' is your DataFrame containing company information
matching_companies = find_companies_by_industry_and_location_and_size_range(data, industry_req, location_req, size_range_req)

# Check for possible values in the dataset to help debug
print("\nAvailable industries in the dataset:")
print(data['industry'].unique())

print("\nAvailable locations in the dataset:")
print(data['locality'].unique())

print("\nAvailable Size's in the dataset:")
print(data['size_range'].unique())

# Display all matching companies
if not matching_companies.empty:
    print("\nCompanies matching your industry and location and size_range requirements:")
    print(matching_companies)
else:
    print("\nNo companies found matching your criteria.")
```
